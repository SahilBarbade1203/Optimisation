```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
fig = plt.figure(figsize=(12,10))
ax = plt.axes(projection="3d")

x = np.linspace(-1,1,60)
y = np.linspace(-1,1,60)    

X, Y = np.meshgrid(x, y)
Z = (X - Y)**4 + 8*X*Y - X + Y + 3

X,Y = np.meshgrid(x,y)
surf = ax.plot_surface(X, Y, Z, cmap = "viridis")

# Set axes label
ax.set_title("Surface plot of Test Fuction",fontsize = 16)
ax.set_xlabel('X1',fontsize = 15 , color = "blue")
ax.set_ylabel('X2',fontsize = 15 , color = "blue")
ax.set_zlabel('Test function',fontsize = 15, color = "blue")


plt.show()
```

```python
y1 = (-1-x)**4 + 8*x*-1 -x -1 +3
z1 = (1-x)**4 + 8*x - x + 1 + 3
axh = plt.figure(figsize=(15,6))
okkh = axh.add_axes([0.1,0.1,0.8,0.8])


okkh.plot(x,z1 , color = "red" ,marker = "x", label = "X2 = 1")
okkh.plot(x,y1,color = "green" ,ls = ":",marker = "o" ,label = "X2 = -1")


okkh.set_title("Test function versus X1 for a fixed X2",fontsize = 16)
okkh.set_xlabel("X1",color= "blue",fontsize = 16)
okkh.set_ylabel("Test Function",color= "blue",fontsize =16)

okkh.legend(loc = 0)
```

```python
okk2 = plt.figure(figsize = (12,6))
X2, Y2 =np.meshgrid(x,y)
Z2 = (X-Y)**4 + 8*X*Y -X +Y +3
ax2 = okk2.add_axes([0.1,0.1,0.8,0.8])
l = ax2.contour(X2,Y2,Z2,20,cmap = "viridis")
plt.clabel(l, inline=True, fontsize=8)

ax2.set_title("Contour plot of test function",fontsize =18)
ax2.set_xlabel("X1",fontsize = 14,color = "blue")
ax2.set_ylabel("X2",fontsize = 14,color = "blue")
```
